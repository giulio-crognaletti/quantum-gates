```python
import pickle as pick
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd

A = pd.read_pickle(r'zne_data_0-100-2-qubits-2-layers')
```

```python
tg = 35 * 10**-9

ZZ = []
ZI = []
IZ = []


N = 0

t = []

for i in range(0,10):
    
    ZZ.append(A[N][i][1]['ZZ'])
    ZI.append(A[N][i][1]['ZI'])
    IZ.append(A[N][i][1]['IZ'])
    t.append(tg * A[0][i][0])

H = np.array(IZ) 


plt.xscale('log')
plt.plot(t,H)
plt.scatter(t,H)
plt.show()
```
